```python
import os
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications import EfficientNetB0  # Simpler architecture
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.callbacks import CSVLogger, EarlyStopping, TensorBoard, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split

print("TensorFlow version:", tf.__version__)

# Define dataset paths
dataset_path = 'E:/Wildfire_Detection_using_UAV_Images_Drone/anomaly/FIRE/THERMAL'  # or RGB
epochs = 18
batch_size = 18  # Adjust based on your GPU memory
n_classes = 3  # Update this to match the number of classes in your dataset
patience = 5

# Load categories
categories = os.listdir(dataset_path)
print("Categories:", categories)

# Select 100 images from each category
selected_files = []
for category in categories:
    category_path = os.path.join(dataset_path, category)
    all_files = os.listdir(category_path)
    selected_files.extend([os.path.join(category_path, file) for file in np.random.choice(all_files, 350, replace=False)])

# Split into train and test sets
train_files, test_files = train_test_split(selected_files, test_size=0.2, random_state=76)

def create_dataframe(file_list):
    data = []
    for file in file_list:
        category = file.split(os.path.sep)[-2]
        data.append((file, category))
    return pd.DataFrame(data, columns=['filename', 'class'])

train_df = create_dataframe(train_files)
test_df = create_dataframe(test_files)

# Image Data Generators
datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    validation_split=0.2,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_generator = datagen.flow_from_dataframe(
    train_df,
    x_col='filename',
    y_col='class',
    target_size=(224, 224),
    color_mode='rgb',
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

validation_generator = datagen.flow_from_dataframe(
    test_df,
    x_col='filename',
    y_col='class',
    target_size=(224, 224),
    color_mode='rgb',
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Model architecture
base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
preds = Dense(n_classes, activation='softmax')(x)
model = Model(inputs=base_model.input, outputs=preds)

# Display model summary
print(model.summary())

# Set all layers to trainable
for layer in model.layers:
    layer.trainable = True

# Compile the Model without mixed precision for better CPU performance
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Setup callbacks
stime = time.strftime("%Y-%m-%d_%H-%M-%S")
log_dir = f"E:/Wildfire_Detection_using_UAV_Images_Drone/anomaly/FIRE/{stime}_{batch_size}b_{epochs}e/logs"
os.makedirs(log_dir, exist_ok=True)

# CSVLogger
csv_log = CSVLogger(f"E:/Wildfire_Detection_using_UAV_Images_Drone/anomaly/FIRE/{stime}_{batch_size}b_{epochs}e/file.csv")

# Early Stopping
early_stopping = EarlyStopping(patience=patience, restore_best_weights=True)

# TensorBoard
tensorboard = TensorBoard(log_dir=log_dir)

# Model Checkpoint
checkpoint_path = os.path.join(log_dir, "cp-{epoch:04d}-{val_accuracy:.4f}va-{val_loss:.4f}vl.keras")
model_checkpoint = ModelCheckpoint(checkpoint_path, save_best_only=True, monitor='val_accuracy', mode='max')

# Reduce Learning Rate on Plateau
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-6)

# Training the Model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=epochs,
    callbacks=[csv_log, early_stopping, tensorboard, model_checkpoint, reduce_lr]
)

# Save the model
model_save_path = f"E:/Wildfire_Detection_using_UAV_Images_Drone/anomaly/FIRE/{stime}_{batch_size}b_{epochs}e/wildfire_detection_model.h5"
model.save(model_save_path)
print(f"Model saved to {model_save_path}")

# Plot training history
model_history = pd.DataFrame(history.history)
model_history.plot()
plt.savefig(f"E:/Wildfire_Detection_using_UAV_Images_Drone/anomaly/FIRE/{stime}_{batch_size}b_{epochs}e/training_history.png")

# Evaluate the Model
evaluation = model.evaluate(validation_generator)
print(f"Evaluation: {evaluation}")

# Image Data Generator for testing
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_generator = test_datagen.flow_from_dataframe(
    test_df,
    x_col='filename',
    y_col='class',
    target_size=(224, 224),
    color_mode='rgb',
    shuffle=False,
    class_mode='categorical',
    batch_size=1
)

activities = test_generator.class_indices
print("Activities:", activities)

def get_activity(val):
    for key, value in activities.items():
        if val == value:
            return key
    return "Invalid"

filenames = test_generator.filenames
nb_samples = len(filenames)
predict = model.predict(test_generator, steps=nb_samples, verbose=1)
y_pred = [get_activity(np.argmax(val)) for val in predict]

y_true = test_df['class'].tolist()
cm = confusion_matrix(y_true, y_pred)
print("Macro Precision:", precision_score(y_true, y_pred, average='macro', zero_division=1))
print("Macro Recall:", recall_score(y_true, y_pred, average='macro', zero_division=1))
print("Macro F1 Score:", f1_score(y_true, y_pred, average='macro', zero_division=1))

print("Micro Precision:", precision_score(y_true, y_pred, average='micro', zero_division=1))
print("Micro Recall:", recall_score(y_true, y_pred, average='micro', zero_division=1))
print("Micro F1 Score:", f1_score(y_true, y_pred, average='micro', zero_division=1))

print("Classification Report:")
print(classification_report(y_true, y_pred))

# Save confusion matrix
dataframe = pd.DataFrame(cm)
inv_dict = {v: k for k, v in activities.items()}
dataframe = dataframe.rename(index=inv_dict)
dataframe = dataframe.rename(columns=inv_dict)
dataframe.to_csv(f"E:/Wildfire_Detection_using_UAV_Images_Drone/anomaly/FIRE/{stime}_{batch_size}b_{epochs}e/Performance_Confusion_Matrix.csv")
```
